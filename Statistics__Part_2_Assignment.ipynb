##Theoretical Questions


1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions based on data.
It involves testing an assumption (hypothesis) about a population parameter.
Based on the evidence, we either reject or fail to reject the hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) states there is no effect or difference.
The alternative hypothesis (H₁) suggests a significant effect or difference exists.
Hypothesis testing aims to determine which hypothesis is more likely.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the threshold for rejecting the null hypothesis, often set at 0.05.
It represents the probability of making a Type I error (false positive).
It helps control the risk of wrongly rejecting a true null hypothesis.

4. What does a P-value represent in hypothesis testing?
- The p-value measures the probability of observing the sample data, assuming
 the null is true.
 A low p-value suggests the data is unlikely under the null hypothesis.
 It helps determine whether to reject the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
- If the p-value ≤ α, reject the null hypothesis (significant result).
 If the p-value > α, fail to reject the null hypothesis (not significant).
 It quantifies the strength of the evidence against the null.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I error: Rejecting a true null hypothesis (false positive).
Type II error: Failing to reject a false null hypothesis (false negative).
Minimizing both is crucial for valid conclusions.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed tests check for an effect in one direction (e.g., greater than).
Two-tailed tests check for an effect in both directions (e.g., not equal).
Choice depends on the hypothesis and research question.

8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test compares sample and population means when population variance is known.
It's used for large sample sizes (n > 30) and normally distributed data.
It determines if the difference is statistically significant.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z = (X - μ) / σ, where X = sample mean, μ = population mean, σ = standard deviation.
The Z-score shows how many standard deviations a value is from the mean.
It helps identify outliers or unusual results.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used when the sample size is small (n < 30) and population standard deviation is unknown.
It is similar to the normal distribution but has heavier tails.
It provides more accurate results for small samples.

11. What is the difference between a Z-test and a T-test?
- Z-test uses known population variance; T-test uses sample variance.
Z-test is for large samples; T-test is for small samples.
Both test differences between means.

12. What is the T-test, and how is it used in hypothesis testing?
- The T-test compares means between two groups to see if they differ significantly.
It’s used when sample size is small and population standard deviation is unknown.
Types include one-sample, independent, and paired T-tests.

13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both are used to compare means and test hypotheses.
Z-test is preferred for large samples with known variance.
T-test is used for small samples with unknown variance.

14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval gives a range of values likely to include the population parameter.
A 95% CI means we’re 95% confident the true value lies within the interval.
It helps assess estimate precision and reliability.

15. What is the margin of error, and how does it affect the confidence interval?
- Margin of error reflects the range of uncertainty around a sample estimate.
A larger margin leads to a wider confidence interval.
It depends on sample size and variability.

16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem updates the probability of a hypothesis based on new evidence.
It’s widely used in predictive modeling and decision-making.
It combines prior and conditional probabilities.

17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used in categorical data analysis.
It assesses the relationship between observed and expected frequencies.
Used in tests like goodness-of-fit and independence.

18. What is the Chi-square goodness of fit test, and how is it applied?
- It tests if observed frequencies match expected ones under a specific hypothesis.
Used to analyze categorical data distribution.
Significant results suggest the data does not fit the expected pattern.

19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution compares two variances and is used in ANOVA tests.
It is skewed and depends on two degrees of freedom.
Common in testing equality of multiple group variances.

20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) compares means across three or more groups.
Assumes independence, normality, and equal variances.
It tells if at least one group differs significantly.

21. What are the different types of ANOVA tests?
- Types include One-Way ANOVA (one factor), Two-Way ANOVA (two factors), and Repeated Measures ANOVA.
Each is suited to different experimental designs.
They all test mean differences across groups.

22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test assesses if group variances are significantly different.
Used in ANOVA and regression analysis.
A significant F-statistic indicates differences between group means or model fits.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

population_mean = 100
sample = np.array([102, 98, 101, 97, 105])
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)

z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 1 - stats.norm.cdf(abs(z_score))

print("1. Z-score:", z_score)
print("P-value:", p_value)
print("Interpretation: ", "Reject null hypothesis" if p_value < 0.05 else "Fail to reject null hypothesis")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)
t_stat, p_val = stats.ttest_1samp(data, 100)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(sample, population_mean, population_std):
    z = (np.mean(sample) - population_mean) / (population_std / np.sqrt(len(sample)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(loc=102, scale=10, size=30)
z, p = one_sample_z_test(sample, 100, 10)

print("Z-score:", z)
print("P-value:", p)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import matplotlib.pyplot as plt

z, _ = one_sample_z_test(sample, 100, 10)
z_critical = stats.norm.ppf(1 - 0.05 / 2)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3)
plt.axvline(z, color='blue', linestyle='--', label=f'Z = {z:.2f}')
plt.title("Two-tailed Z-test Decision Region")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_errors(mu0, mu1, sigma, n, alpha=0.05):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    z_alpha = stats.norm.ppf(1 - alpha)
    critical_value = mu0 + z_alpha * (sigma / np.sqrt(n))
    beta = stats.norm.cdf((critical_value - mu1) / (sigma / np.sqrt(n)))

    plt.plot(x, stats.norm.pdf(x, mu0, sigma/np.sqrt(n)), label="H0")
    plt.plot(x, stats.norm.pdf(x, mu1, sigma/np.sqrt(n)), label="H1")
    plt.axvline(critical_value, color='red', linestyle='--', label="Critical value")
    plt.fill_between(x, stats.norm.pdf(x, mu0, sigma/np.sqrt(n)), where=(x >= critical_value), alpha=0.3, color='red', label="Type I Error")
    plt.fill_between(x, stats.norm.pdf(x, mu1, sigma/np.sqrt(n)), where=(x < critical_value), alpha=0.3, color='green', label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.grid(True)
    plt.show()
    return beta

beta = visualize_errors(100, 105, 15, 30)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)
    margin = stats.t.ppf((1 + confidence) / 2, len(sample)-1) * sem
    return mean - margin, mean + margin

ci = confidence_interval(sample)
print("Confidence Interval:", ci)
print("Interpretation: We are 95% confident the true mean lies within this interval.")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy import stats

data = np.random.normal(50, 10, 100)
confidence = 0.95
n = len(data)
std_err = stats.sem(data)
margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err
print("Margin of Error:", margin)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_A, prior_notA, likelihood_B_given_A, likelihood_B_given_notA):
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + likelihood_B_given_notA * prior_notA
    return numerator / denominator

posterior = bayes_theorem(0.01, 0.99, 0.9, 0.1)
print("Posterior Probability:", posterior)
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

observed = np.array([20, 30, 50])
total = np.sum(observed)

# Let's say we expect equal distribution among 3 categories
expected_proportions = np.array([1/3, 1/3, 1/3])
expected = expected_proportions * total  # Scale to match observed total exactly

chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples.
group1 = np.random.normal(10, 2, 30)
group2 = np.random.normal(10, 4, 30)

f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
df1, df2 = len(group1)-1, len(group2)-1
p_value = 1 - stats.f.cdf(f_stat, df1, df2)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    for i, g in enumerate(groups):
        stat, p = shapiro(g)
        print(f"Group {i+1} Normality p-value:", p)
    stat, p = levene(*groups)
    print("Levene’s Test for equal variances p-value:", p)

check_anova_assumptions(group1, group2, group3)
```
